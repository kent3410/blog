```python
# ベイズの識別境界の図示

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

x = np.linspace(-10,10,10000)

ax.plot(x, (1/2) * norm.pdf(x, loc = -1, scale = 1), color="red", linestyle="solid", linewidth=1.0, label="P(x | C1) P(C1)")
ax.plot(x, (1/2) * norm.pdf(x, loc = 1, scale = 1), color="blue", linestyle="solid", linewidth=1.0, label="P(x | C2) P(C2)")
ax.axvline(0, ls="--", color="green")

ax.set_title("Bayes Boundary")
ax.set_xlabel("x")
ax.grid(True)
fig.legend()
fig.show()
fig.savefig("bayes_boundary.png")
```
